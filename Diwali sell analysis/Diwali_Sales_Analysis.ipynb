## EDA with python of Diwali Sales dataset

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#loading the data set
df=pd.read_csv('/content/drive/MyDrive/Projects/EDA with python/Diwali_Sales_Analysis/Diwali Sales Data.csv',encoding='unicode-escape')
```

```python
#checking data of the data set
df.head()
```

```python
#finding number of rows and columns
df.shape
```

We can see that there are 11251 rows and 15 columns

```python
#finding info of the data set
df.info()
```

I found that status and unnamed1 has no data and amount column has some missing values

```python
#deleting no data columns
df1=df.drop(['Status','unnamed1'],axis=1)
```

```python
#dataset after deleting two columns
df1.head()
```

```python
#finding null values
pd.isnull(df1).sum()
```

I found that only Amount column has 12 missing value

```python
#Handling missing values
df1.dropna(inplace=True)
```

```python
#changing the data type of Amount column from float to int
df1['Amount']=df1['Amount'].astype('int')
```

```python
#replacing 0 with unmarried and 1 with married customer
df1['Marital_Status'].replace([0,1],['Married','Unmarried'],inplace=True)
```

```python
df1['Marital_Status'].replace(['Unmarried','Married'],['Married','Unmarried'],inplace=True)
```

```python
#Look out of our data set
df1.head()
```

```python
#showing info
df1.info()
```

we can see now we have a final data set which have total 13 columns and 11239 rows and there is no missing value 


```python
#showing 13 column names
df1.columns
```

```python
#finding some numeric insighs
df1[['Age','Amount','Orders']].describe()
```

So, we found that average age of our customers are 35, average amount 9453 rupees and average orders 2.48.
Customers minimum age of 12 and maximum age 92.
Minimum amount 188 rupees and maxmimum amount 23952 rupees.
Mimimum orders 1 and maximum orders 4.

# Performing Exploratory Data Analysis

```python
df1['Cust_name'].value_counts().head(5)
```

```python
sns.barplot(x=list(df1['Cust_name'].value_counts().head(5).keys()),y=list(df1['Cust_name'].value_counts().head(5)))
plt.show()
```

We can see that these are the top 5 customers who bought most time.

```python
df1['Cust_name'].value_counts().tail(5)
```

```python
sns.barplot(x=list(df1['Cust_name'].value_counts().tail(5).keys()),y=list(df1['Cust_name'].value_counts().tail(5)))
plt.show()
```

We can see that these are the 5 customers who bought less time.

```python
leb=sns.countplot(x='Gender',data=df1)
for i in leb.containers:
  leb.bar_label(i)
```

Between male and female, female customers are most, approximately twice of male customers

```python
leb=sns.countplot(x='Age Group',data=df1,hue='Gender')
for i in leb.containers:
  leb.bar_label(i)
```

Age group 26-35 bought most, then followed by 36-45,18-25,46-50 and 51-55.

```python
gender_amount=df1.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
plot_gender_amount=sns.barplot(x='Gender', y='Amount' ,data=gender_amount)
for i in plot_gender_amount.containers:
  plot_gender_amount.bar_label(i)
```

Women customers also spend maximum amount rather than male customer.

```python
sns.set(rc={'figure.figsize':(10,5)})
Marital_Status_Amount=df1.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=Marital_Status_Amount,hue='Gender')
plt.show()
```

Married female customer bought most

```python
state_amount=df1.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.pie(state_amount['Amount'].head(),labels=state_amount['State'].head(),autopct="%0.1f%%")
plt.show()
```

So, here are top 5 states with-respect to Amount of purchase

```python
state_order=df1.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
plt.pie(state_order['Orders'].head(),labels=state_amount['State'].head(),autopct="%0.1f%%")
plt.show()
```

So, here are top 5 states with-respect to Order number

```python
Zone_order=df1.groupby(['Zone'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(x=Zone_order['Zone'],y=Zone_order['Orders'])
plt.show()
```

Central region order most and eastern region order least

```python
Zone_amount=df1.groupby(['Zone'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x=Zone_amount['Zone'],y=Zone_amount['Amount'])
plt.show()
```

As like orders, central zone come first with-respect to amount of purchase

```python
sns.set(rc={'figure.figsize':(21,5)})
occupation_plot=sns.countplot(x='Occupation' ,data=df1)
for i in occupation_plot.containers:
  occupation_plot.bar_label(i)
```

Customer belong from It sector has bought most product,then followed by healthcare, aviation, banking occupation and govt. employee.

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x=list(df1['Product_Category'].value_counts().head(5).keys()),y=list(df1['Product_Category'].value_counts().head(5)))
plt.show()
```

We can see that Clothing and Apparel is the highest sell category

```python
sns.set(rc={'figure.figsize':(15,5)})
product_amount=df1.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Amount',data=product_amount)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,5)})
product_order=df1.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Orders',data=product_order)
plt.show()
```
